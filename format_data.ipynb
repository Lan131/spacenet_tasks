```python
import geojson
import shutil
import os
from os import listdir
from os.path import isfile, join
import numpy as np
```

```python

path_geojson="./processedData/processedBuildingLabels/vectordata/geojson/"
onlygeojson = [f for f in listdir(path_geojson) if isfile(join(path_geojson, f))]

path_tif="./processedData/processedBuildingLabels/3band/"
onlytif = [f for f in listdir(path_tif) if isfile(join(path_tif, f))]
```

```python
len(onlygeojson)==len(onlytif)
```

```python
with open("./processedData/processedBuildingLabels/vectordata/geojson/Geo_AOI_1_RIO_img1.geojson") as f:
    gj = geojson.load(f)
gj
```

```python
k=0
for geojson_file in onlygeojson:
    with open(path_geojson+geojson_file) as f:
        gj = geojson.load(f)
        
    if 'building' in str(gj):
        shutil.move(path_tif+str(onlytif[k]), "./processedData/buildings/"+str(onlytif[k]))
    else:
        shutil.move(path_tif+str(onlytif[k]), "./processedData/no_buildings/"+str(onlytif[k]))
    k=k+1    
    
```

```python
#Match test data labels
```

```python
path_tif="./src/test/"
onlytif = [f for f in listdir(path_tif) if isfile(join(path_tif, f))]
```

```python
results=[]
for tif_file in onlytif:
    geojson_file="Geo_"+str(tif_file)[6:-4]+".geojson"
    #print(tif_file)
    #print(path_geojson+geojson_file)
    with open(path_geojson+geojson_file) as f:
        gj = geojson.load(f)
        
    if 'building' in str(gj):
        results.append(geojson_file+"-building")
    else:
        results.append(geojson_file+"-none")
```

```python
results
```

```python
with open("testing.txt", "w") as output:
    output.write(str(results))
```
